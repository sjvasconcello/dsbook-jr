```python
import statistics
import random

# Supongamos que tienes la lista datos_aleatorios con los 10,000 datos generados previamente
datos_aleatorios = np.random.normal(loc=1, scale=0.5, size=10000)
datos_aleatorios = np.clip(datos_aleatorios, 0, 2)
datos_aleatorios = list(datos_aleatorios)  # Convertir a lista

# Extraer muestra de 100 datos y calcular la media
muestra_100 = random.sample(datos_aleatorios, 100)
media_100 = statistics.mean(muestra_100)
print("Muestra de 100 datos - Media:", media_100)

# Extraer muestra de 500 datos y calcular la media
muestra_500 = random.sample(datos_aleatorios, 500)
media_500 = statistics.mean(muestra_500)
print("Muestra de 500 datos - Media:", media_500)

# Extraer muestra de 1000 datos y calcular la media
muestra_1000 = random.sample(datos_aleatorios, 1000)
media_1000 = statistics.mean(muestra_1000)
print("Muestra de 1000 datos - Media:", media_1000)

# Extraer muestra de 5000 datos y calcular la media
muestra_5000 = random.sample(datos_aleatorios, 5000)
media_5000 = statistics.mean(muestra_5000)
print("Muestra de 5000 datos - Media:", media_5000)
```

```python
import matplotlib.pyplot as plt

# Supongamos que tienes las muestras muestra_100, muestra_500, muestra_1000, muestra_5000

# Configuración de los parámetros de las distribuciones normales
mu_100, sigma_100 = np.mean(muestra_100), np.std(muestra_100)
mu_500, sigma_500 = np.mean(muestra_500), np.std(muestra_500)
mu_1000, sigma_1000 = np.mean(muestra_1000), np.std(muestra_1000)
mu_5000, sigma_5000 = np.mean(muestra_5000), np.std(muestra_5000)

# Generar datos para las distribuciones normales
x_100 = np.linspace(mu_100 - 3*sigma_100, mu_100 + 3*sigma_100, 100)
x_500 = np.linspace(mu_500 - 3*sigma_500, mu_500 + 3*sigma_500, 500)
x_1000 = np.linspace(mu_1000 - 3*sigma_1000, mu_1000 + 3*sigma_1000, 1000)
x_5000 = np.linspace(mu_5000 - 3*sigma_5000, mu_5000 + 3*sigma_5000, 5000)

# Calcular las densidades de probabilidad para las distribuciones normales
y_100 = 1/(sigma_100 * np.sqrt(2*np.pi)) * np.exp(-(x_100 - mu_100)**2 / (2*sigma_100**2))
y_500 = 1/(sigma_500 * np.sqrt(2*np.pi)) * np.exp(-(x_500 - mu_500)**2 / (2*sigma_500**2))
y_1000 = 1/(sigma_1000 * np.sqrt(2*np.pi)) * np.exp(-(x_1000 - mu_1000)**2 / (2*sigma_1000**2))
y_5000 = 1/(sigma_5000 * np.sqrt(2*np.pi)) * np.exp(-(x_5000 - mu_5000)**2 / (2*sigma_5000**2))

# Crear el gráfico
plt.figure(figsize=(15, 5))

plt.plot(x_100, y_100, label='Muestra de 100 datos')
plt.plot(x_500, y_500, label='Muestra de 500 datos')
plt.plot(x_1000, y_1000, label='Muestra de 1000 datos')
plt.plot(x_5000, y_5000, label='Muestra de 5000 datos')

plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribuciones normales de las muestras')
plt.legend()

plt.show()
```
